Matplotlib & Seaborn Plotting Content

Univariate Analysis --> The operations / characteristics involves with single feature / variable.
Bivariate Analysis --> The operations / Characteristics invloves with 2 features / 2 variables.
Multi-variate Analysis --> The Analysis involves more than 2 variables.

Univariate Analysis :
Numerical 
(i) Histogram
(ii) Distribution plot
(iii) Boxplot
Categorical 
(i) Bar Chart
(ii) Pie Chart

Bi-Variate Analysis : 
Numerical - Numerical :
(i) Scatterplot
(ii) Line Plot
Numerical - Categorical :
(i) Box plot
(ii) Bar plot
(iii) Swarm plot
(iv) Strip plot
(v) Viloin Plot
Categorical - Categorical:
(i) Count Plot

Multi-Variate Analysis : 
(i) Pairplot
(ii) Heatmap

SUBPLOTS

In subplot first we have to define the structure and then we are making a plot over that
fig --> stores the figure and its the name of the structure
axes --> it specifies where we wanted to plot in the canvas and stores the number of rows and columns in axes
width and height should be also defined

```python
fig,axes=plt.subplots(1,2,figsize=(15,5))
data['sex'].value_counts().plot(kind='pie',ax=axes[0])
data['sex'].value_counts().plot(kind='bar',ax=axes[1])
plt.show()
```

```python
fig,axes=plt.subplots(2,1,figsize=(15,5))
data['day'].value_counts().plot(kind='pie',ax=axes[0])
data['day'].value_counts().plot(kind='bar',ax=axes[1])
plt.show()
```
